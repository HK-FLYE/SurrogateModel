```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.font_manager import FontProperties
import matplotlib as mpl

mpl.rcParams['font.family'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False

# 假设数据
categories = ['过度磨损', '严重变形', '疲劳断裂', '表面破损']
correct_predictions = [16, 18, 19, 17]  # 正确预测的数量
incorrect_predictions = [3, 3, 3, 1]  # 错误预测的数量

# 设置柱状图中每一组柱子的位置
index = np.arange(len(categories))

# 绘制柱状图
fig, ax = plt.subplots()
bar_heights = [correct + incorrect for correct, incorrect in zip(correct_predictions, incorrect_predictions)]

# 绘制正确预测的柱状图部分
bars_correct = ax.bar(index, correct_predictions, label='正确', color='#ADD8E6')

# 在正确预测的基础上堆叠错误预测的柱状图部分
bars_incorrect = ax.bar(index, incorrect_predictions, bottom=correct_predictions, label='错误', color='#800080')

# 添加图表细节
ax.set_xlabel('故障类型')
ax.set_ylabel('预测数量')
ax.set_title('')
ax.set_xticks(index)
ax.set_xticklabels(categories)
plt.legend(loc='best')

# 显示图表
plt.show()
```
