```python
import matplotlib.pyplot as plt
import numpy as np

# 假设数据
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
correct_predictions = [10, 15, 12, 8]  # 正确预测的数量
incorrect_predictions = [10, 5, 8, 12]  # 错误预测的数量

# 设置柱状图中每一组柱子的位置
index = np.arange(len(categories))

# 绘制柱状图
fig, ax = plt.subplots()
bar_heights = [correct + incorrect for correct, incorrect in zip(correct_predictions, incorrect_predictions)]

# 绘制正确预测的柱状图部分
bars_correct = ax.bar(index, correct_predictions, label='Correct', color='#ADD8E6')

# 在正确预测的基础上堆叠错误预测的柱状图部分
bars_incorrect = ax.bar(index, incorrect_predictions, bottom=correct_predictions, label='Incorrect', color='#800080')

# 添加图表细节
ax.set_xlabel('Categories')
ax.set_ylabel('Number of Predictions')
ax.set_title('Prediction Results by Category')
ax.set_xticks(index)
ax.set_xticklabels(categories)
ax.legend()

# 显示图表
plt.show()
```
